```python
import pandas as pd
import numpy as np
from geopy.geocoders import Nominatim
import requests
from bs4 import BeautifulSoup

url = "https://en.wikipedia.org/wiki/Armorial_of_Europe"

response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

table = soup.find('table', {'class': 'wikitable'})
print(table)
rows = table.find_all('tr')

countries = []
coats = []
blazons = []
mottos = []
articles = []
data = {}

for row in rows:
    columns = row.find_all('td')
    if columns:
        country = columns[0].text.strip()
        coat_of_arms = columns[1].find('img')['src'] if columns[1].find('img') else None
        blazon = columns[2].text.strip()
        motto_text = columns[3].text.strip()
        article = columns[4].text.strip()
        countries.append(country)
        coats.append(coat_of_arms)
        blazons.append(blazon)
        mottos.append(motto_text)
        articles.append(article)
#countries ={"countries":countries}
#coats = {"coats":coats}
#blazons = {"blazons":blazons}
#mottos = {"mottos":mottos}
#articles = {"articles" : articles}   
data.update({"countries":countries,"coats":coats,"blazons":blazons,"mottos":mottos,"articles":articles} )
        

        #print(f"Country: {country}")
        #print(f"Coat of Arms URL: https://en.wikipedia.org{coat_of_arms}")
        #print(f"Blazon: {blazon}")
        #print(f"Motto/Text: {motto_text}")
        #print(f"Article: {article}")
        #print("-------------------")
```

```python
def get_flag_url(country):

    country = country.replace(' ', '_')


    url = f'https://en.wikipedia.org/wiki/{country}'
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')


    infobox = soup.find('table', {'class': 'infobox'})
    if infobox == None:
        return np.nan
    else:
        image = infobox.find('img')


        flag_url = 'https:' + image['src']

        return flag_url
```
